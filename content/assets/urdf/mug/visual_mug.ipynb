```python
import matplotlib.pyplot as plt
from pytransform3d.urdf import UrdfTransformManager
from mpl_toolkits.mplot3d import Axes3D
```

```python

# 创建变换管理器的实例
tm = UrdfTransformManager()

# 从文件中加载URDF模型
with open("movable_collision_test.urdf", 'r') as urdf_file:
    urdf_string = urdf_file.read()
tm.load_urdf(urdf_string)
```

```python
# 创建一个图形和一个3D轴来绘制
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
```

```python
ax = tm.plot_frames_in(
    "base", s=0.1,
    show_name=False)
```

```python
tm.plot_visuals("base", ax=ax, convex_hull_of_mesh=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 创建一个新的图像和轴
fig, ax = plt.subplots()

# 示例数据
x = range(10)
y = range(10)

# 绘制数据
ax.plot(x, y)

# 设置x轴刻度步长
ax.set_xticks(range(0, 10, 2))  # 每隔2个单位显示一个刻度

# 设置y轴刻度步长
ax.set_yticks(range(0, 10, 2))  # 每隔2个单位显示一个刻度

# 控制网格密度
ax.grid(True)  # 打开网格
ax.xaxis.set_major_locator(plt.MultipleLocator(3))  # 控制x轴主要刻度的间隔
ax.yaxis.set_major_locator(plt.MultipleLocator(3))  # 控制y轴主要刻度的间隔

# 显示图形
plt.show()
```
